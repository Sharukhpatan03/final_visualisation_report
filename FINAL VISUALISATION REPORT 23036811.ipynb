```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load the dataset
data = pd.read_csv('climate_change_indicators.csv')
```

```python
data.head(50)
```

```python
# Function to create a histogram/bar chart/pie chart
def create_histogram(data):
    plt.figure(figsize=(10, 6))
    sns.histplot(data['F2022'], kde=True)
    plt.title('Distribution of Temperature Change in 2022')
    plt.xlabel('Temperature Change (°C)')
    plt.ylabel('Frequency')
    plt.show()
    
# Call the functions to create the plots
create_histogram(data)
```

```python
# Function to create a line/scatter graph
def create_line_graph(data):
    plt.figure(figsize=(10, 6))
    plt.plot(data['Country'], data['F2013'], marker='o', label='2013')
    plt.plot(data['Country'], data['F2022'], marker='o', label='2022')
    plt.xticks(rotation=90)
    plt.title('Temperature Change (°C) Over Time (2013-2022)')
    plt.xlabel('Country')
    plt.ylabel('Temperature Change (°C)')
    plt.legend()
    plt.show()
    
create_line_graph(data.head(10))
```

```python
# Function to create a bar chart
def create_bar_chart(data):
    plt.figure(figsize=(10, 6))
    top_countries = data.sort_values(by='F2022', ascending=False).head(10)
    plt.bar(top_countries['Country'], top_countries['F2022'])
    plt.title('Top 10 Countries with Highest Temperature Change in 2022')
    plt.xlabel('Country')
    plt.ylabel('Temperature Change (°C)')
    plt.xticks(rotation=45, ha='right')
    plt.show()
    
# Call the functions to create the plots
create_bar_chart(data)
```

```python
# Function to create a pie chart
def create_pie_chart(data):
    plt.figure(figsize=(8, 8))
    total_temperature_change = data['F2022'].sum()
    top_countries = data.sort_values(by='F2022', ascending=False).head(5)
    plt.pie(top_countries['F2022'], labels=top_countries['Country'], autopct='%1.1f%%', startangle=140)
    plt.title('Distribution of Temperature Change by Top 5 Countries in 2022')
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.show()
create_pie_chart(data)
```

```python
# Function to create a scatter plot
def create_scatter_plot(data):
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x='F2013', y='F2022', data=data)
    plt.title('Scatter Plot of Temperature Change (2013 vs 2022)')
    plt.xlabel('Temperature Change in 2013 (°C)')
    plt.ylabel('Temperature Change in 2022 (°C)')
    plt.show()

# Call the functions to create the plots
create_scatter_plot(data)
```

```python
# Function to create a box plot
def create_box_plot(data):
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='F2022', data=data)
    plt.title('Box Plot of Temperature Change in 2022')
    plt.xlabel('Temperature Change in 2022 (°C)')
    plt.show()
create_box_plot(data)
```

```python
def create_violin_plot(data):
    plt.figure(figsize=(10, 6))
    sns.violinplot(x='F2022', data=data)
    plt.title('Violin Plot of Temperature Change in 2022')
    plt.xlabel('Temperature Change in 2022 (°C)')
    plt.show()

create_violin_plot(data)
```

```python
def create_joint_plot(data):
    sns.set(style="white", color_codes=True)
    g = sns.jointplot(x='F2013', y='F2022', data=data, kind='scatter', height=7)
    g.set_axis_labels('Temperature Change in 2013 (°C)', 'Temperature Change in 2022 (°C)', fontsize=12)
    plt.subplots_adjust(top=0.9)
    plt.suptitle('Joint Plot of Temperature Change (2013 vs 2022)', fontsize=16)
    plt.show()

create_joint_plot(data)
```

```python
def create_pair_plot(data):
    sns.pairplot(data[['F2013', 'F2022']], kind='scatter', diag_kind='kde', height=3)
    plt.suptitle('Pair Plot of Temperature Change (2013 vs 2022)', y=1.02)
    plt.show()

create_pair_plot(data)
```

```python
def create_heatmap(data):
    plt.figure(figsize=(10, 8))
    sns.heatmap(data[['F2013', 'F2022']].corr(), annot=True, cmap='coolwarm', fmt=".2f")
    plt.title('Correlation Heatmap of Temperature Change (2013 vs 2022)')
    plt.show()

create_heatmap(data)
```
